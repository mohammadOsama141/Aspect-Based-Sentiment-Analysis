# Train data preprocessing

```python
import csv
import xml.etree.ElementTree as ET
import pandas as pd
```

## Conversion from XML to CSV


```python
#XML file
tree = ET.parse('/content/train.xml')
root = tree.getroot()


with open('MAMS_train.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['ID', 'Review', 'Aspects', 'Sentiment']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    #iteration through data
    for idx, sentence in enumerate(root.findall('sentence')):
        review_text = sentence.find('text').text

        #extracting aspect categories and polarities
        aspects = []
        polarities = []
        for aspectCategory in sentence.find('aspectCategories').findall('aspectCategory'):
            aspects.append(aspectCategory.attrib['category'])
            polarities.append(aspectCategory.attrib['polarity'])

        writer.writerow({'ID': idx + 1,
                         'Review': review_text,
                         'Aspects': ', '.join(aspects),
                         'Sentiment': ', '.join(polarities)})
```

```python
train = pd.read_csv("/content/MAMS_train.csv")
```

```python
#saving csv
train.to_csv('/content/MAMS_train.csv', index=False)
```
